# Mission Corridors

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
from shapely.geometry import LineString
from datetime import date
import cpi

#pd.set_option('display.max_rows', 1000) 
#pd.set_option('display.max_columns', 1000)
#pd.set_option("display.max_colwidth", None) 
```

```python


# 1. Load your existing geojson
gdf = gpd.read_file("docs/Mission/mission_recovery.geojson")

# 2. Corridor LINESTRINGS (using your coordinates)

sixteenth = LineString([
    (-122.42639848568001, 37.76463787050363),   # Dolores & 16th
    (-122.41862447318765, 37.76513520717117)    # Capp & 16th
])

twentyfourth = LineString([
    (-122.42070428208872, 37.752113850440224),  # Valencia & 24th
    (-122.40636144939559, 37.75298767165918)    # Potrero & 24th
])

valencia = LineString([
    (-122.4223937835193, 37.76987633443966),    # Valencia & Duboce
    (-122.42030618333321, 37.74813359059528)    # Valencia & Cesar Chavez
])

# 3. Combine into a single MULTILINE corridor object
corridor_lines = gpd.GeoSeries([sixteenth, twentyfourth, valencia], crs="EPSG:4326")
corridor_union = corridor_lines.unary_union   # MultiLineString

# 4. Select polygons that touch/intersect the corridor line (NO BUFFER)
#    Using `intersects` catches even tiny touches.
touching = gdf[gdf.geometry.intersects(corridor_union)]

# 5. Save output
touching.to_file("docs/Mission/mission_corridors_touching.geojson", driver="GeoJSON")

print("DONE — created mission_corridors_touching.geojson")
```
