# Mission Corridors

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
from shapely.geometry import LineString
from datetime import date
import cpi

#pd.set_option('display.max_rows', 1000) 
#pd.set_option('display.max_columns', 1000)
#pd.set_option("display.max_colwidth", None) 
```

```python
# 1. Load your existing Mission geojson
gdf = gpd.read_file("docs/Mission/mission_recovery.geojson")

# 2. Define corridor centerlines using YOUR coordinates

sixteenth = LineString([
    (-122.42639848568001, 37.76463787050363),   # Dolores & 16th
    (-122.41862447318765, 37.76513520717117)    # Capp & 16th
])

twentyfourth = LineString([
    (-122.42070428208872, 37.752113850440224),  # Valencia & 24th
    (-122.40636144939559, 37.75298767165918)    # Potrero & 24th
])

valencia = LineString([
    (-122.4223937835193, 37.76987633443966),    # Valencia & Duboce
    (-122.42030618333321, 37.74813359059528)    # Valencia & Cesar Chavez
])

# 3. Create a GeoSeries of the corridor lines
corridors = gpd.GeoSeries([sixteenth, twentyfourth, valencia], crs="EPSG:4326")

# 4. Buffer the lines
# 60 meters is usually a perfect "commercial corridor" width
corridor_buffer = corridors.to_crs(3857).buffer(1).to_crs(4326)

# 5. Merge all buffers into one shape
merged_buffer = corridor_buffer.unary_union

# 6. Select polygons that intersect the corridor buffer
filtered = gdf[gdf.intersects(merged_buffer)]

# 7. Save result
filtered.to_file("docs/Mission/mission_corridors.geojson", driver="GeoJSON")

print("DONE! Created mission_corridors.geojson")
```
