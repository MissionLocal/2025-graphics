```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('/Users/agneslu/Desktop/neighborhood-data/data-new/housing-tenure/ACSDP5Y2023.DP04-Data.csv')
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# drop the last column, the header is NaN
df = df.drop(df.columns[-1], axis=1)
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df.head(3)
```

```python
# only keep the columns of Area, Toal, and columns with HOUSING TENURE in it
df = df[['Area'] + ['Estimate!!HOUSING OCCUPANCY!!Total housing units'] + [col for col in df.columns if 'HOUSING TENURE' in col]]
df.head(3)
```

```python
# keep the first four columns
columns_to_keep = ['Area', 'Estimate!!HOUSING OCCUPANCY!!Total housing units', 'Estimate!!HOUSING TENURE!!Occupied housing units', 'Estimate!!HOUSING TENURE!!Occupied housing units!!Owner-occupied', 'Estimate!!HOUSING TENURE!!Occupied housing units!!Renter-occupied']
df = df[columns_to_keep]
df.head(3)
```

```python
# rename columns
df = df.rename(columns={
    'Estimate!!HOUSING OCCUPANCY!!Total housing units': 'Total',
    'Estimate!!HOUSING TENURE!!Occupied housing units': 'Occupied',
    'Estimate!!HOUSING TENURE!!Occupied housing units!!Owner-occupied': 'Owner occupied',
    'Estimate!!HOUSING TENURE!!Occupied housing units!!Renter-occupied': 'Renter occupied'
})
df.head(3)
```

```python
# change values to numeric
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')
df['Occupied'] = pd.to_numeric(df['Occupied'], errors='coerce')
df['Owner occupied'] = pd.to_numeric(df['Owner occupied'], errors='coerce')
df['Renter occupied'] = pd.to_numeric(df['Renter occupied'], errors='coerce')
```

```python
# add a column for vacant housing units
df['Vacant'] = df['Total'] - df['Occupied']
df.head(3)
```

```python
# only keep columns we need
columns_to_keep = ['Area', 'Total', 'Owner occupied', 'Renter occupied', 'Vacant']
df = df[columns_to_keep]
df.head(3)
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Total', 'Owner occupied', 'Renter occupied', 'Vacant']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    'owner-occupied': df.set_index('Area')['Owner occupied'].to_dict(),
    'renter-occupied': df.set_index('Area')['Renter occupied'].to_dict(), 
    'vacant': df.set_index('Area')['Vacant'].to_dict(),
    'total': df.set_index('Area')['Total'].to_dict()
}

# convert to JSON and save to file
import json

with open('/Users/agneslu/Desktop/neighborhood-data/data-new/housing_tenure.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'housing_tenure.json'")
```
