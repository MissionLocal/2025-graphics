```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('/Users/agneslu/Desktop/neighborhood-data/data-new/disability/ACSDT5Y2023.B18101-Data.csv')
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# drop the last column, the header is NaN
df = df.drop(df.columns[-1], axis=1)
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df
```

```python
# ignore the totals and only look at disability breakdowns
df = df.filter(regex=r'Area|disability', axis=1)
df
```

```python
# convert all columns except 'Area' to numeric
numeric_columns = [col for col in df.columns if col != 'Area']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# fill any NaN values with 0
df[numeric_columns] = df[numeric_columns].fillna(0)
df
```

```python
# get columns that contain "With a disability" and "No disability"
with_disability_cols = [col for col in df.columns if 'With a disability' in col]
no_disability_cols = [col for col in df.columns if 'No disability' in col]

# calculate totals for each row
df['Total_With_Disability'] = df[with_disability_cols].sum(axis=1)
df['Total_No_Disability'] = df[no_disability_cols].sum(axis=1)
df['Total_Population'] = df['Total_With_Disability'] + df['Total_No_Disability']
```

```python
# new df
df = df[['Area', 'Total_With_Disability', 'Total_No_Disability', 'Total_Population']]
df
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Total_With_Disability', 'Total_No_Disability', 'Total_Population']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    'disabled': df.set_index('Area')['Total_With_Disability'].to_dict(),
    'not disabled': df.set_index('Area')['Total_No_Disability'].to_dict(), 
    'total': df.set_index('Area')['Total_Population'].to_dict()
}

# convert to JSON and save to file
import json

with open('/Users/agneslu/Desktop/neighborhood-data/data-new/disability.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'disability.json'")
```
