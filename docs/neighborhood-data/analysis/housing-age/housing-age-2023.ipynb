```python
import pandas as pd
```

```python
# import data
df = pd.read_csv('/Users/agneslu/Desktop/neighborhood-data/data-new/housing-age/ACSDP5Y2023.DP04-Data.csv')
df.head(3)
```

```python
# drop the first column
df = df.drop(df.columns[0], axis=1)
df.head(3)
```

```python
# change NAME to Area
df = df.rename(columns={'NAME': 'Area'})
df.head(3)
```

```python
# drop the margin columns
df = df[df.columns.drop(list(df.filter(regex='M')))]
df.head(3)
```

```python
# drop the last column, the header is NaN
df = df.drop(df.columns[-1], axis=1)
df.head(3)
```

```python
# rename columns
column_mapping = {}
for i, col in enumerate(df.columns):
    if i < len(df.iloc[0]):  # Make sure we don't go out of bounds
        column_mapping[col] = df.iloc[0, i]

df = df.rename(columns=column_mapping)
df.head(3)
```

```python
# remove the first row
df = df.drop(df.index[0])  
df = df.reset_index(drop=True)
df.head(3)
```

```python
# change Geographic Area Name to Area
df = df.rename(columns={'Geographic Area Name': 'Area'})
```

```python
# only keep numbers for the area column
df['Area'] = df['Area'].str.replace(r'.*Census Tract\s+', '', regex=True)  # remove everything before tract number
df['Area'] = df['Area'].str.replace(r';.*', '', regex=True)  # remove everything after semicolon
df['Area'] = df['Area'].str.strip()  # remove any extra whitespace
df.head(3)
```

```python
# only keep the column of Area, and columns with YEAR STRUCTURE BUILT in it
df = df[['Area'] + [col for col in df.columns if 'YEAR STRUCTURE BUILT' in col]]
df.head(3)
```

```python
# get rid of Estimate!!YEAR STRUCTURE BUILT!!Total housing units!! in the column names
df.columns = df.columns.str.replace('Estimate!!YEAR STRUCTURE BUILT!!Total housing units!!', '', regex=True)
df.head(3)
```

```python
# change the second column name to Total
df = df.rename(columns={df.columns[1]: 'Total'})
df.head(3)
```

```python
# drop columns with Percent!! in it
df = df[df.columns.drop(list(df.filter(regex='Percent!!')))]
df.head(3)
```

```python
# combine a few columns
df['2000 to current'] = df[['Built 2020 or later', 'Built 2010 to 2019', 'Built 2000 to 2009']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['1970 to 2000'] = df[['Built 1990 to 1999', 'Built 1980 to 1989', 'Built 1970 to 1979']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['1940 to 1970'] = df[['Built 1960 to 1969', 'Built 1950 to 1959', 'Built 1940 to 1949']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Pre-1940'] = df[['Built 1939 or earlier']].apply(pd.to_numeric, errors='coerce').sum(axis=1)
# only keep the columns we need
columns_to_keep = ['Area', 'Total', '2000 to current', '1970 to 2000', '1940 to 1970', 'Pre-1940']
df = df[columns_to_keep]
df.head(3)
```

```python
# convert numeric columns to proper data types
numeric_columns = ['Total', '2000 to current', '1970 to 2000', '1940 to 1970', 'Pre-1940']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# create the dictionary in the format you want
result_dict = {
    '2000 to current': df.set_index('Area')['2000 to current'].to_dict(),
    '1970 to 2000': df.set_index('Area')['1970 to 2000'].to_dict(), 
    '1940 to 1970': df.set_index('Area')['1940 to 1970'].to_dict(),
    'pre-1940': df.set_index('Area')['Pre-1940'].to_dict(),
    'total': df.set_index('Area')['Total'].to_dict()
}

# convert to JSON and save to file
import json

with open('/Users/agneslu/Desktop/neighborhood-data/data-new/housing_age.json', 'w') as f:
    json.dump(result_dict, f, indent=2)

print("JSON file saved as 'housing_age.json'")
```
